For this assingment, I created an edge detection algorithm using the canny algorithm. 

Lets take an example image. Here's what it looks like:

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(plt.imread("messi.jpeg"))
plt.axis("off")
plt.show()
```

The first step of the algorithm is to convert the image to grayscale:

```python
plt.imshow(plt.imread("messi_grey.jpeg"))
plt.axis("off")
plt.show()
```

The next step is to use a gaussian blur to smooth the image.

```python
plt.imshow(plt.imread("messi_blurred.jpeg"))
plt.axis("off")
plt.show()
```

Next, we use a sobell operator to calculate the image gradients (magintude and direction). Below is an image of just the gradient magnitudes.

```python
plt.imshow(plt.imread("messi_sobel.jpeg"))
plt.axis("off")
plt.show()
```

Next we have to non maximum suppression and then hysteresis to thin out the edges and finalize edge detection.

```python
plt.imshow(plt.imread("messi_hysteresized.jpeg"))
plt.axis("off")
plt.show()
```

At the start, there were a few challenges I had to face when figuring out how much to blur the image, experimenting with the kernel size and the sigma for the gaussian distribution, so that the edges were clear in the final image, but eventually I landed on a kernel of size = 5 and sigma = 2.
